```python
import sqlite3, pandas as pd
conn = sqlite3.connect("boston.db")

# Check total rows
print(pd.read_sql("SELECT COUNT(*) AS rows FROM violations", conn))

# Peek at first 5 rows
print(pd.read_sql("SELECT * FROM violations LIMIT 5", conn))

# Quick check: top violation codes
print(pd.read_sql("""
    SELECT violation_code, COUNT(*) AS cnt
    FROM violations
    GROUP BY violation_code
    ORDER BY cnt DESC
    LIMIT 10
""", conn))
```
